```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
CryptoClose = pd.read_csv("CryptoClose.csv",index_col=[0],parse_dates=[0])
```

```python
CryptoClose
```

```python
CryptoClose.plot(figsize=(15,8),fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
CryptoClose.iloc[0]
```

```python
norm=CryptoClose.div(CryptoClose.iloc[0]).mul(100)
```

```python
norm
```

```python
norm.plot(figsize=(15,8),fontsize=13)
plt.legend(fontsize=13)
plt.show()
```
